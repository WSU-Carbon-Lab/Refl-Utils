```python
class AllParams(pd.DataFrame):
    cols = [
        "Transition Energy (eV)",
        "Transition Amplitude (a.u.)",
        "Transition Width (eV)",
        "TDM_x",
        "TDM_y",
        "TDM_z",
        "TDM Theta (degrees)",
        "Transition Symmetry",
        "TDM_xx",
        "TDM_yy",
        "TDM_zz",
        "TDM_xy",
        "TDM_xz",
        "TDM_yz",
        "Originating atom",
        "Originating MO",
        #  "Final Cluster transition was placed in",
    ]

    def __init__(self, filepath):
        super().__init__()
        df = pd.read_csv(filepath, skiprows=1)
        df.columns = self.cols
        self.__dict__.update(df.__dict__)


dft = AllParams(d_path / "allParamsSorted.txt")
sns.lmplot(
    data=dft,
    palette="pastel",
    x="Transition Energy (eV)",
    y="TDM Theta (degrees)",
    fit_reg=False,
)
dft
```

```python
class AngleResNexafs(pd.DataFrame):
    def __init__(
        self,
        filepath: str | Path,
        molecular_name: str,
        angles=["90", "70", "55", "40", "20"],
        *args,
        **kwargs,
    ):
        df = pd.read_csv(filepath, dtype="float64[pyarrow]")
        cols = df.columns
        en_cols = [col for col in cols if "Energy" in col]
        df = df.iloc[1:-2]
        df["Energy [eV]"] = df[en_cols].mean(axis=1)
        df.drop(en_cols, axis=1, inplace=True)
        df.index = df["Energy [eV]"]
        df.drop("Energy [eV]", axis=1, inplace=True)
        df.rename(
            columns={col: angles[i] for i, col in enumerate(df.columns)}, inplace=True
        )

        self.__dict__.update(df.__dict__)
        self.molecular_name = molecular_name
        self.density = kwargs.get("density", 2.3)
        self.normalize(molecular_name, *args, **kwargs)

    def normalize(self, molecular_name, density=2.3):
        if r"$\beta_{ba}$" and r"$\beta_{ba}$" in self.columns:
            energies = self.index.values * 10 ** (-3)
            bare_atom = xsf.index_of_refraction(
                molecular_name, energy=energies, density=density
            )
            self[r"$\beta_{ba}$"] = -bare_atom.imag
            self[r"$\delta_{ba}$"] = 1 - bare_atom.real

        if r"$\delta_{zz}" in self.columns:
            # normalize delta

            self[r"$\delta_{zz}$"] /= (
                self[r"$\delta_{zz}$"].iloc[0] / self[r"$\delta_{ba}$"].iloc[0]
            )
            self[r"$\delta_{xx}$"] /= (
                self[r"$\delta_{xx}$"].iloc[0] / self[r"$\delta_{ba}$"].iloc[0]
            )

        else:
            # normalize beta

            norm_factor = self["55"].iloc[0] / self[r"$\delta_{ba}$"]
            self /= norm_factor

    def plot_diffspec(self):
        try:
            fig, ax = plt.subplots(
                nrows=2, sharex=True, gridspec_kw={"height_ratios": [3, 1], "hspace": 0}
            )

            self.plot(
                ax=ax[0], y=["55", "20", "90"], colors=sns.color_palette("coolwarm", 3)
            )
            self.plot(ax=ax[1], y="Diff", legend=False)
            ax[0].set(
                ylabel="Nexafs [a.u.]",
                title="ZnPc Optical Constants",
                ticklabel_format="sci",
            )
            ax[1].set(ylabel="Diff", xlabel="Energy [eV]")
        except:
            print("Generate the difference spectrum first using df.get_diffspec()")

    def get_diffspec(self, plot=True):
        self["Diff"] = self["20"] - self["90"]
        if plot:
            self.plot_diffspec()

    def plot_oocs(self):
        try:
            fig, ax = plt.subplots(
                nrows=3,
                sharex=True,
                gridspec_kw={"height_ratios": [2, 3, 1], "hspace": 0},
            )
            self.plot(ax=ax[0], y=["55", "20", "90"])
            self.plot(ax=ax[2], y="Diff", legend=False)
            self.plot(ax=ax[1], y=[r"$\beta_{zz}$", r"$\beta_{xx}$"])
            ax[0].set(ylabel="Nexafs [a.u.]", title="ZnPc Optical Constants")
            ax[1].set(ylabel="Optical Constants [a.u.]")
            ax[2].set(ylabel="Diff [a.u.]", xlabel="Energy [eV]")
        except:
            print("Generate the out of plane spectrum first using df.get_ooc()")

    def get_ooc(self, plot=True):
        if "Diff" not in self.columns:
            self.get_diffspec(plot=False)

        iso_intensity = self["55"].max()
        diff_intensity = self["Diff"].max()
        self[r"$\beta_{zz}$"] = self["55"]
        self[r"$\beta_{xx}$"] = self["55"]

        zsf = 2 * iso_intensity / diff_intensity
        xsf = (iso_intensity) / diff_intensity

        self[r"$\beta_{zz}$"] += zsf * self["Diff"]
        self[r"$\beta_{xx}$"] -= xsf * self["Diff"]

        # kramers kronig transform
        beta_zz = np.column_stack(
            (self.index.to_numpy(), self[[r"$\beta_{zz}$"]].to_numpy())
        )
        beta_xx = np.column_stack(
            (self.index.to_numpy(), self[[r"$\beta_{xx}$"]].to_numpy())
        )

        # kkcalc weirdness
        kk_kwargs = {
            "ChemicalFormula": self.molecular_name,
            "load_options": None,
            "input_data_type": "Beta",
            "merge_points": [self.index[0], self.index[-1]],
            "fix_distortions": False,
            "curve_tolerance": 0.05,
            "curve_recursion": 100,
        }
        self.formula_mass = data.calculate_FormulaMass(
            kk.data.ParseChemicalFormula(self.molecular_name)
        )

        zz_sf = kk.kk_calculate_real(beta_zz, **kk_kwargs)[0, [0, 1]]
        xx_sf = kk.kk_calculate_real(beta_xx, **kk_kwargs)[0, [0, 1]]

        self[r"$\delta_{zz}$"] = zz_sf[:, 1][
            self.index[0] <= zz_sf[:, 0] <= self.index[-1]
        ]
        self[r"$\delta_{xx}$"] = xx_sf[:, 1][
            self.index[0] <= xx_sf[:, 0] <= self.index[-1]
        ]

        self.normalize(self.molecular_name, density=self.density)

    def plot_ar(self):
        dft.plot(y=self.columns[:4], ylabel="Nexafs [a.u]", title="ZnPC Nexafs")
```

```python
en = [250, 283.7]


def get_delta_beta(n):
    delta = 1 - n.real
    beta = -n.imag
    return delta + 1j * beta


_n_znpc = xsf.index_of_refraction(
    "C32H16N8Zn", density=2.2, energy=[e * 1e-3 for e in en]
)
n_znpc = get_delta_beta(_n_znpc)
print(f"The isotropic refractive index of ZnPc is {n_znpc}")
```

```python
""" 
Parameter Constraints
"""

max_thick = 300
max_rough = 10

sio2_t = 8.000003
sio2_rough = 6.38308
sio2_rho = 1.62331

carbon_thick = 6.08467
carbon_rough = 4.40733

znpc_thick = 200
znpc_rough = 0

vac = [PXR_MaterialSLD("", 1)(0, 0) for e in en]
si = [PXR_MaterialSLD("Si", 2.33, energy=en, name="Si")(0, 1.5) for e in en]
sio2 = [PXR_MaterialSLD("SiO2", sio2_rho, name="SiO2")(sio2_t, sio2_rough) for e in en]
carbon = [
    PXR_MaterialSLD("C", 1.8, name="Carbon")(carbon_thick, carbon_rough) for e in en
]

znpc = np.zeros(2, dtype=object)
znpc[0] = PXR_MaterialSLD("C32H16N8Zn", 2.3, energy=en[0], name="ZnPc")(
    znpc_thick, znpc_rough
)
znpc[1] = PXR_SLD()

structure = []
for i, _ in enumerate(en):
    struct = vac | znpc[i] | carbon | sio2 | si
    struct.plot()
    structure.append(struct)
```

```python
from Logp import LogpExtra_rough

scale_vary = False
scale_lb = 0.6
scale_ub = 1.2

bkg_vary = False
bkg_lb = 1e-9
bkg_ub = 9e-6

random_state = 1
nwalkers = 100
nsamples = 200

move = [(DEMove(sigma=1e-7), 0.95), (DEMove(sigma=1e-7), 0.05)]
chain_name = "znpc_room_temp.txt"
save_fitter = "znpc_room_temp_fitter.pkl"

model = PXR_ReflectModel(
    structure, scale=1, bkg=0, dq=0, energy=en, pol="sp", name=f"ZnPc_{en}"
)

model.scale.setp(vary=scale_vary, bounds=(scale_lb, scale_ub))
model.bkg.setp(vary=bkg_vary, bounds=(bkg_lb, bkg_ub))

objective = Objective(model, znpc_data, transform=Transform("logY"), name=f"obj_{en}")
lpe = LogpExtra_rough(objective)
objective.logp_extra = lpe
objective.plot()
```

```python
sio2.thick.setp(vary=False)
sio2.rough.setp(vary=False)

carbon.thick.setp(vary=False)
carbon.rough.setp(vary=False)
carbon.sld.density.setp(vary=True, bounds=(1.5, 2))

znpc.thick.setp(vary=True, bounds=(max_thick / 2, max_thick))
znpc.rough.setp(vary=True, bounds=(0, max_rough))
znpc.sld.density.setp(vary=True, bounds=(1.5, 3))
```

```python
fitter = CurveFitter(objective, nwalkers=nwalkers, moves=move)
fitter.initialise("jitter")
chain = fitter.sample(int(nsamples), nthin=10, random_state=random_state)
lp = fitter.logpost
plt.plot(-lp, yscale="log")

print(objective.varying_parameters())
objective.plot()
structure.plot()
```

```python
with open(d_path / save_fitter, "wb") as f:
    pickle.dump(fitter, f)

with open(d_path / chain_name, "wb") as f:
    f.write(str(objective.parameters).encode())
```
